```python
!pip install rdkit
import pandas as pd
from rdkit import Chem
```

```python


# Standardize SMILES strings
def standardize_smiles(smiles):
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is not None:
            # Remove salts
            mol = Chem.RemoveHs(mol)
            Chem.SanitizeMol(mol)

            # Standardize stereochemistry (if needed)
            # ... add stereochemistry standardization logic here ...

            return Chem.MolToSmiles(mol)
        else:
            return None  # Or handle invalid SMILES as needed
    except Exception as e:
        print(f"Error standardizing SMILES '{smiles}': {e}")
        return None

df['standardized_smiles'] = df['canonical_smiles'].apply(standardize_smiles)
df.head(3)
```

```python


import numpy as np

# Assuming 'standard_value' column contains IC50 values
# Convert IC50 to pIC50
def convert_IC50_to_pIC50(IC50):
    pIC50 = -np.log10(IC50 * 10**-9)  # Assuming IC50 is in nM
    return pIC50

# Convert 'standard_value' column to numeric type
df['standard_value'] = pd.to_numeric(df['standard_value'], errors='coerce')

# Apply the conversion to create a new column
df['pIC50'] = df['standard_value'].apply(lambda x: convert_IC50_to_pIC50(x) if x > 0 else np.nan)

# Remove rows with missing pIC50 values
df = df.dropna(subset=['pIC50'])

df.head()
```

```python
from rdkit.Chem import Descriptors

# Calculate molecular weight
df['MW'] = df['standardized_smiles'].apply(lambda x: Descriptors.MolWt(Chem.MolFromSmiles(x)) if x is not None else None)

# Calculate LogP
df['LogP'] = df['standardized_smiles'].apply(lambda x: Descriptors.MolLogP(Chem.MolFromSmiles(x)) if x is not None else None)

# Calculate number of hydrogen bond donors
df['HBD'] = df['standardized_smiles'].apply(lambda x: Descriptors.NumHDonors(Chem.MolFromSmiles(x)) if x is not None else None)

# Calculate number of hydrogen bond acceptors
df['HBA'] = df['standardized_smiles'].apply(lambda x: Descriptors.NumHAcceptors(Chem.MolFromSmiles(x)) if x is not None else None)

# Calculate TPSA
df['TPSA'] = df['standardized_smiles'].apply(lambda x: Descriptors.TPSA(Chem.MolFromSmiles(x)) if x is not None else None)

# Calculate number of rotatable bonds
df['RotatableBonds'] = df['standardized_smiles'].apply(lambda x: Descriptors.NumRotatableBonds(Chem.MolFromSmiles(x)) if x is not None else None)


# Display the updated dataframe with new features
df.head(3)
```

```python


import numpy as np
# Import necessary libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Define features (X) and target variable (y)
X = df[['MW', 'LogP', 'HBD', 'HBA', 'TPSA', 'RotatableBonds']]
y = df['pIC50']

# Convert pIC50 to binary classes (inhibitor/non-inhibitor)
# You need to define a threshold for pIC50 to classify as inhibitor or not
threshold = 5 # Example threshold - adjust as needed
y_binary = np.where(y >= threshold, 1, 0) # 1 for inhibitor, 0 for non-inhibitor

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.2, random_state=42)

# Initialize and train the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42) # Adjust hyperparameters as needed
rf_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = rf_classifier.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of Random Forest Classifier: {accuracy}")
```

```python
from sklearn.metrics import roc_auc_score, precision_score, recall_score, f1_score

# Make predictions on the test set
y_pred = rf_classifier.predict(X_test)
y_pred_proba = rf_classifier.predict_proba(X_test)[:, 1] # Probabilities for the positive class


# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
auc_roc = roc_auc_score(y_test, y_pred_proba)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(f"AUC-ROC: {auc_roc}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")
```
